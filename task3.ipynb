# dealing with missing values

```python
# since in features brand, model, and age there is only 1 nan value so handling them with simple imputation
# for numerical with mean
# age is having numerical data, so replacing the 1 nan value in age with the mean of all valuse in age
dataset['age'].fillna(dataset['age'].mean(), inplace=True)
```

```python
# for categorical feature with mode(most frequent value)
# for brand
dataset['brand'].fillna(dataset['brand'].mode()[0], inplace=True)
```

```python
# for model
dataset['model'].fillna(dataset['model'].mode()[0], inplace=True)
```

# distance should be in float but it is in object, so fixing that

```python
dataset['Distance'] = dataset['Distance'].str.replace(' Meters','')
```

```python
# coverting all into meters and removing the string part 'meters'
def distance(x) :
    p = x.split()
    try :
        if p[1] == 'KM' :
            return (str(round(float(p[0])*1000,1)))
        elif p[1] == 'Miles' :
            return (str(round(float(p[0])*1000*1609,1)))
    except :
        return x
```

```python
dataset['Distance'] = dataset['Distance'].apply(distance)
```

```python
dataset['Distance'] = dataset['Distance'].astype('float64')
```

```python
dataset = dataset.rename(columns = {'Distance':'Distance(Meters)'})
```
